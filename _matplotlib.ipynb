```python
#A continuacion importaremos dos bibliotecas: numpy para cálculos numéricos y pyplot para graficar
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Creamos una variable x con un conjunto de valores igualmente espaciados
x = np.linspace(0, 6.28)

#Para este conjunto de valores, calculamos sen(x)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
#Si has usado matlab antes, pyplot tiene comandos similares para graficar
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfica de y = sen(x)"); #<-- Porqué he puesto este punto y coma final?
```
